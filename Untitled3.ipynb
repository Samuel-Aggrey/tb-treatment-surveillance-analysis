## Load and Preprocess Data

### Subtask:
Load the 'TB_outcomes_2025-12-13.csv' dataset into a pandas DataFrame, perform initial data cleaning, handle missing values, and prepare the data for analysis. This step will include parsing date columns and ensuring correct data types for epidemiological calculations.


**Reasoning**:
First, I will load the 'TB_outcomes_2025-12-13.csv' dataset into a pandas DataFrame and display its first few rows and information to understand its structure and data types.



```python
import pandas as pd

df = pd.read_csv('/content/TB_outcomes_2025-12-13.csv')

print('First 5 rows of the DataFrame:')
print(df.head())

print('\nDataFrame Info:')
df.info()
```

**Reasoning**:
I will inspect the unique values and their counts for categorical columns like 'country', 'iso2', 'iso3', 'g_whoregion', and 'year' to identify any inconsistencies or typos. This helps in understanding the data quality and preparing for further cleaning.



```python
print('\nUnique values and counts for "country":')
print(df['country'].value_counts())

print('\nUnique values and counts for "iso2":')
print(df['iso2'].value_counts())

print('\nUnique values and counts for "iso3":')
print(df['iso3'].value_counts())

print('\nUnique values and counts for "g_whoregion":')
print(df['g_whoregion'].value_counts())

print('\nUnique values and counts for "year":')
print(df['year'].value_counts().sort_index())
```

**Reasoning**:
I will check the percentage of missing values for all columns to understand the extent of missing data and inform the strategy for handling them.



```python
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing_values, 'Missing Percentage': missing_percentage})
print('Missing values percentage per column:')
print(missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False))
```

**Reasoning**:
I will handle missing values by filling the few missing `iso2` values with 'Unknown' and imputing the majority of numerical columns (which represent counts or outcomes) with 0. This approach addresses missing data while retaining as much information as possible for epidemiological calculations, as `NaN` often implies zero reported cases in such datasets.



```python
df['iso2'] = df['iso2'].fillna('Unknown')

# Get a list of numerical columns that are of float type and have missing values
numerical_cols_with_nan = df.select_dtypes(include=['float64']).columns

# Fill NaN values in these numerical columns with 0
for col in numerical_cols_with_nan:
    df[col] = df[col].fillna(0)

print('Number of missing values after imputation:')
print(df.isnull().sum().sum())
```

**Reasoning**:
I will create a 'date' column from the 'year' column to facilitate time-series analysis and ensure consistency for epidemiological calculations, although the 'year' column is already of an appropriate integer type.



```python
df['date'] = pd.to_datetime(df['year'], format='%Y')

print('DataFrame Info after adding date column:')
df.info()
print('\nFirst 5 rows with new date column:')
print(df[['year', 'date']].head())
```

## Calculate Epidemiological Indicators

### Subtask:
Compute key epidemiological indicators such as treatment success rate, mortality rate, and treatment failure rate for each country and year based on the preprocessed data.


**Reasoning**:
I will calculate the 'Treatment Success Rate (new_sp)', 'Mortality Rate (new_sp)', and 'Treatment Failure Rate (new_sp)' columns as per the instructions, ensuring to handle division by zero errors by setting the rate to 0 when the cohort is zero. Then, I will display the first few rows of the DataFrame including these new columns to verify the calculations.



```python
df['Treatment Success Rate (new_sp)'] = ((df['new_sp_cur'] + df['new_sp_cmplt']) / df['new_sp_coh'] * 100).fillna(0)
df.loc[df['new_sp_coh'] == 0, 'Treatment Success Rate (new_sp)'] = 0

df['Mortality Rate (new_sp)'] = (df['new_sp_died'] / df['new_sp_coh'] * 100).fillna(0)
df.loc[df['new_sp_coh'] == 0, 'Mortality Rate (new_sp)'] = 0

df['Treatment Failure Rate (new_sp)'] = (df['new_sp_fail'] / df['new_sp_coh'] * 100).fillna(0)
df.loc[df['new_sp_coh'] == 0, 'Treatment Failure Rate (new_sp)'] = 0

print('First 5 rows with new epidemiological indicators:')
print(df[['country', 'year', 'new_sp_coh', 'new_sp_cur', 'new_sp_cmplt', 'new_sp_died', 'new_sp_fail',
          'Treatment Success Rate (new_sp)', 'Mortality Rate (new_sp)', 'Treatment Failure Rate (new_sp)']].head())
```

## Design Early Warning Alert System

### Subtask:
Implement an automated alert mechanism that detects unusual increases in TB cases using statistical methods.


**Reasoning**:
I will calculate the 3-year rolling mean and standard deviation of 'new_sp_coh' for each country. These rolling statistics will be used to establish a baseline for detecting unusual increases in TB cases.



```python
df['rolling_mean_3yr'] = df.groupby('country')['new_sp_coh'].rolling(window=3, min_periods=1).mean().reset_index(level=0, drop=True)
df['rolling_std_3yr'] = df.groupby('country')['new_sp_coh'].rolling(window=3, min_periods=1).std().reset_index(level=0, drop=True)

print("First 5 rows with rolling mean and std:")
print(df[['country', 'year', 'new_sp_coh', 'rolling_mean_3yr', 'rolling_std_3yr']].head())
```

**Reasoning**:
I will define an 'alert' condition where 'new_sp_coh' exceeds the 'rolling_mean_3yr' by two times the 'rolling_std_3yr', create a new boolean column 'alert' based on this condition, and then display the rows where an alert is present.



```python
df['alert'] = (df['new_sp_coh'] > (df['rolling_mean_3yr'] + 2 * df['rolling_std_3yr']))

alert_countries = df[df['alert']][['country', 'year', 'new_sp_coh', 'rolling_mean_3yr', 'rolling_std_3yr', 'alert']]

print('Countries with an alert for unusual increase in TB cases:')
print(alert_countries.head())
```

## Integrate Forecasting Module

### Subtask:
Develop and integrate a time-series forecasting model to predict future TB case trends for the next three years.


**Reasoning**:
I will aggregate the `new_sp_coh` column by `year` to create a time series of total new TB cases per year. I will then convert this aggregated data into a DataFrame with columns 'ds' (year as datetime) and 'y' (total new cases), which is the required format for the Prophet model.



```python
total_tb_cases_per_year = df.groupby('year')['new_sp_coh'].sum().reset_index()
total_tb_cases_per_year.columns = ['year', 'new_sp_coh']

prophet_df = pd.DataFrame({
    'ds': pd.to_datetime(total_tb_cases_per_year['year'], format='%Y'),
    'y': total_tb_cases_per_year['new_sp_coh']
})

print('Prophet DataFrame head:')
print(prophet_df.head())
print('\nProphet DataFrame info:')
prophet_df.info()
```

**Reasoning**:
I will import the Prophet class, instantiate and fit the model using the prepared `prophet_df`, then create a future DataFrame for three years and generate predictions. Finally, I will display the relevant prediction columns to show the forecasted trends.



```python
import prophet

# Instantiate and fit the Prophet model
m = prophet.Prophet(yearly_seasonality=True)
m.fit(prophet_df)

# Create a DataFrame for future predictions (3 years / 3 periods as data is yearly)
future = m.make_future_dataframe(periods=3, freq='Y')

# Generate predictions
forecast = m.predict(future)

print('First 5 rows of the forecast:')
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```

**Reasoning**:

```python
import streamlit as st
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
from prophet import Prophet # Import Prophet for forecasting

def load_and_preprocess_data(file_path=None, uploaded_file=None):
    """Loads, cleans, and preprocesses the TB outcomes data."""
    if uploaded_file:
        df_new = pd.read_csv(uploaded_file)
    elif file_path:
        df_new = pd.read_csv(file_path)
    else:
        raise ValueError("Either file_path or uploaded_file must be provided.")

    # Handle missing iso2 values
    df_new['iso2'] = df_new['iso2'].fillna('Unknown')

    # Fill NaN values in numerical columns with 0
    numerical_cols_with_nan = df_new.select_dtypes(include=['float64']).columns
    for col in numerical_cols_with_nan:
        df_new[col] = df_new[col].fillna(0)

    # Create 'date' column
    df_new['date'] = pd.to_datetime(df_new['year'], format='%Y')

    # Calculate epidemiological indicators
    df_new['Treatment Success Rate (new_sp)'] = ((df_new['new_sp_cur'] + df_new['new_sp_cmplt']) / df_new['new_sp_coh'] * 100).fillna(0)
    df_new.loc[df_new['new_sp_coh'] == 0, 'Treatment Success Rate (new_sp)'] = 0

    df_new['Mortality Rate (new_sp)'] = (df_new['new_sp_died'] / df_new['new_sp_coh'] * 100).fillna(0)
    df_new.loc[df_new['new_sp_coh'] == 0, 'Mortality Rate (new_sp)'] = 0

    df_new['Treatment Failure Rate (new_sp)'] = (df_new['new_sp_fail'] / df_new['new_sp_coh'] * 100).fillna(0)
    df_new.loc[df_new['new_sp_coh'] == 0, 'Treatment Failure Rate (new_sp)'] = 0

    # For early warning system
    df_new['rolling_mean_3yr'] = df_new.groupby('country')['new_sp_coh'].rolling(window=3, min_periods=1).mean().reset_index(level=0, drop=True)
    df_new['rolling_std_3yr'] = df_new.groupby('country')['new_sp_coh'].rolling(window=3, min_periods=1).std().reset_index(level=0, drop=True)
    df_new['alert'] = (df_new['new_sp_coh'] > (df_new['rolling_mean_3yr'] + 2 * df_new['rolling_std_3yr']))

    return df_new

# Initialize session state for the DataFrame if not already present
if 'df' not in st.session_state:
    st.session_state['df'] = load_and_preprocess_data(file_path='/content/TB_outcomes_2025-12-13.csv')
    print("Default data loaded and preprocessed into session_state['df']")

df = st.session_state['df']

# Re-calculate derived dataframes using the current df in session state
total_tb_cases_per_year = df.groupby('year')['new_sp_coh'].sum().reset_index()
total_tb_cases_per_year.columns = ['year', 'new_sp_coh']

average_treatment_success_rate_per_year = df.groupby('year')['Treatment Success Rate (new_sp)'].mean().reset_index()
average_treatment_success_rate_per_year.columns = ['year', 'Average Treatment Success Rate (new_sp)']

alert_countries = df[df['alert']][['country', 'year', 'new_sp_coh', 'rolling_mean_3yr', 'rolling_std_3yr', 'alert']]

# For forecasting
prophet_df = pd.DataFrame({
    'ds': pd.to_datetime(total_tb_cases_per_year['year'], format='%Y'),
    'y': total_tb_cases_per_year['new_sp_coh']
})

# For global risk map
tb_burden_by_country = df.groupby(['country', 'iso3'])['new_sp_coh'].sum().reset_index()
tb_burden_by_country.columns = ['country', 'iso3', 'total_new_sp_coh']

# For research insights
mdr_tb_by_country = df.groupby('country')['mdr_coh'].sum().sort_values(ascending=False)
mortality_rate_by_country = df.groupby('country')['Mortality Rate (new_sp)'].mean().sort_values(ascending=False)

st.set_page_config(layout="wide")

st.title('Tuberculosis Surveillance Dashboard')

# Sidebar for navigation
st.sidebar.title('Navigation')
page_selection = st.sidebar.radio(
    'Go to',
    [
        'Global Dashboard',
        'Country Analysis',
        'Early Warning System',
        'Forecasting',
        'Global Risk Map',
        'Research Insights',
        'Data Management',
        'Automated Report Generation' # Added new option
    ]
)

# Main content area based on selection
if page_selection == 'Global Dashboard':
    st.header('Global Dashboard - Aggregated Trends')
    st.write('### Total New TB Cases Per Year')
    st.line_chart(total_tb_cases_per_year.set_index('year'))

    st.write('### Average Treatment Success Rate Per Year')
    st.line_chart(average_treatment_success_rate_per_year.set_index('year'))

elif page_selection == 'Country Analysis':
    st.header('Country-Level Detailed Analysis')

    unique_countries = df['country'].unique()
    # Ensure selected_country is defined for the initial load if it was from previous cell execution
    # If not, set a default or use the first unique country.
    if 'selected_country' not in locals() and 'selected_country' not in globals():
        selected_country = unique_countries[0]

    selected_country_analysis = st.selectbox(
        'Select a Country for Detailed Analysis:',
        unique_countries,
        index=list(unique_countries).index(selected_country) if selected_country in unique_countries else 0
    )

    country_analysis_df = df[df['country'] == selected_country_analysis].copy()

    if not country_analysis_df.empty:
        st.write(f'### TB Trends and Epidemiological Indicators in {selected_country_analysis}')
        fig_country, ax_country = plt.subplots(figsize=(12, 6))
        ax_country.plot(country_analysis_df['year'], country_analysis_df['new_sp_coh'], label='New TB Cases (Cohort)', marker='o')
        ax_country.plot(country_analysis_df['year'], country_analysis_df['Treatment Success Rate (new_sp)'], label='Treatment Success Rate (%)', marker='x')
        ax_country.plot(country_analysis_df['year'], country_analysis_df['Mortality Rate (new_sp)'], label='Mortality Rate (%)', marker='s')
        ax_country.plot(country_analysis_df['year'], country_analysis_df['Treatment Failure Rate (new_sp)'], label='Treatment Failure Rate (%)', marker='d')

        ax_country.set_title(f'TB Trends and Epidemiological Indicators in {selected_country_analysis} Over Time')
        ax_country.set_xlabel('Year')
        ax_country.set_ylabel('Value')
        ax_country.legend()
        ax_country.grid(True)
        plt.xticks(rotation=45)
        st.pyplot(fig_country)
    else:
        st.write('No data available for the selected country.')

elif page_selection == 'Early Warning System':
    st.header('Early Warning System')
    st.write('### Countries with Unusual Increase in TB Cases')
    if not alert_countries.empty: # Check if alert_countries is not empty
        st.dataframe(alert_countries)
    else:
        st.write('No unusual increases detected based on current criteria.')

elif page_selection == 'Forecasting':
    st.header('Forecasting Module')
    st.write('### Predicted Future TB Trends')

    m = Prophet(yearly_seasonality=True)
    m.fit(prophet_df)
    future = m.make_future_dataframe(periods=3, freq='YE')
    forecast_prophet = m.predict(future)

    fig1_st = m.plot(forecast_prophet)
    plt.title('TB Cases Forecast')
    plt.xlabel('Year')
    plt.ylabel('New TB Cases Cohort')
    st.pyplot(fig1_st)

    fig2_st = m.plot_components(forecast_prophet)
    st.pyplot(fig2_st)

elif page_selection == 'Global Risk Map':
    st.header('Global Risk Map')
    st.write('### Global TB Burden by Country')
    fig_map = px.choropleth(
        tb_burden_by_country,
        locations='iso3',
        color='total_new_sp_coh',
        hover_name='country',
        color_continuous_scale=px.colors.sequential.Plasma,
        title='Global TB Burden by Country'
    )
    st.plotly_chart(fig_map)

elif page_selection == 'Research Insights':
    st.header('Automated Research Insights')
    st.write('### Top 5 Countries with Highest Total MDR-TB Cohort')
    st.dataframe(mdr_tb_by_country.head())

    st.write('### Top 5 Countries with Highest Average Mortality Rate (new_sp)')
    st.dataframe(mortality_rate_by_country.head())

elif page_selection == 'Data Management': # New section for data management
    st.header('Data Management')
    st.write('Upload a new dataset or reset to the default dataset.')

    uploaded_file = st.file_uploader("Upload your TB surveillance CSV file", type=["csv"])

    if uploaded_file is not None:
        try:
            st.session_state['df'] = load_and_preprocess_data(uploaded_file=uploaded_file)
            st.success("File successfully uploaded and processed!")
            st.experimental_rerun() # Rerun to update dashboard with new data
        except Exception as e:
            st.error(f"Error processing uploaded file: {e}")

    if st.button('Reset to Default Data'):
        st.session_state['df'] = load_and_preprocess_data(file_path='/content/TB_outcomes_2025-12-13.csv')
        st.success("Dataset reset to default!")
        st.experimental_rerun() # Rerun to update dashboard with default data

elif page_selection == 'Automated Report Generation': # New section for report generation
    st.header('Automated Report Generation')
    st.write('Generate a downloadable report for a selected country.')

    unique_countries_report = df['country'].unique()
    selected_country_report = st.selectbox(
        'Select a Country for Report Generation:',
        unique_countries_report
    )

    report_df = df[df['country'] == selected_country_report].copy()

    if not report_df.empty:
        # Select relevant columns for the report
        report_data = report_df[['year', 'new_sp_coh', 'Treatment Success Rate (new_sp)',
                                   'Mortality Rate (new_sp)', 'Treatment Failure Rate (new_sp)']]
        st.write(f'### Epidemiological Indicators for {selected_country_report}')
        st.dataframe(report_data)

        # Provide download button for CSV
        csv_data = report_data.to_csv(index=False).encode('utf-8')
        st.download_button(
            label="Download Report (CSV)",
            data=csv_data,
            file_name=f"{selected_country_report}_TB_Report.csv",
            mime="text/csv",
        )
    else:
        st.write('No data available for the selected country to generate a report.')
```

```python
import streamlit as st
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
from prophet import Prophet # Import Prophet for forecasting

def load_and_preprocess_data(file_path=None, uploaded_file=None):
    """Loads, cleans, and preprocesses the TB outcomes data."""
    if uploaded_file:
        df_new = pd.read_csv(uploaded_file)
    elif file_path:
        df_new = pd.read_csv(file_path)
    else:
        raise ValueError("Either file_path or uploaded_file must be provided.")

    # Handle missing iso2 values
    df_new['iso2'] = df_new['iso2'].fillna('Unknown')

    # Fill NaN values in numerical columns with 0
    numerical_cols_with_nan = df_new.select_dtypes(include=['float64']).columns
    for col in numerical_cols_with_nan:
        df_new[col] = df_new[col].fillna(0)

    # Create 'date' column
    df_new['date'] = pd.to_datetime(df_new['year'], format='%Y')

    # Calculate epidemiological indicators
    df_new['Treatment Success Rate (new_sp)'] = ((df_new['new_sp_cur'] + df_new['new_sp_cmplt']) / df_new['new_sp_coh'] * 100).fillna(0)
    df_new.loc[df_new['new_sp_coh'] == 0, 'Treatment Success Rate (new_sp)'] = 0

    df_new['Mortality Rate (new_sp)'] = (df_new['new_sp_died'] / df_new['new_sp_coh'] * 100).fillna(0)
    df_new.loc[df_new['new_sp_coh'] == 0, 'Mortality Rate (new_sp)'] = 0

    df_new['Treatment Failure Rate (new_sp)'] = (df_new['new_sp_fail'] / df_new['new_sp_coh'] * 100).fillna(0)
    df_new.loc[df_new['new_sp_coh'] == 0, 'Treatment Failure Rate (new_sp)'] = 0

    # For early warning system
    df_new['rolling_mean_3yr'] = df_new.groupby('country')['new_sp_coh'].rolling(window=3, min_periods=1).mean().reset_index(level=0, drop=True)
    df_new['rolling_std_3yr'] = df_new.groupby('country')['new_sp_coh'].rolling(window=3, min_periods=1).std().reset_index(level=0, drop=True)
    df_new['alert'] = (df_new['new_sp_coh'] > (df_new['rolling_mean_3yr'] + 2 * df_new['rolling_std_3yr']))

    return df_new

# Initialize session state for the DataFrame if not already present
if 'df' not in st.session_state:
    st.session_state['df'] = load_and_preprocess_data(file_path='/content/TB_outcomes_2025-12-13.csv')
    print("Default data loaded and preprocessed into session_state['df']")

df = st.session_state['df']

# Re-calculate derived dataframes using the current df in session state
total_tb_cases_per_year = df.groupby('year')['new_sp_coh'].sum().reset_index()
total_tb_cases_per_year.columns = ['year', 'new_sp_coh']

average_treatment_success_rate_per_year = df.groupby('year')['Treatment Success Rate (new_sp)'].mean().reset_index()
average_treatment_success_rate_per_year.columns = ['year', 'Average Treatment Success Rate (new_sp)']

alert_countries = df[df['alert']][['country', 'year', 'new_sp_coh', 'rolling_mean_3yr', 'rolling_std_3yr', 'alert']]

# For forecasting
prophet_df = pd.DataFrame({
    'ds': pd.to_datetime(total_tb_cases_per_year['year'], format='%Y'),
    'y': total_tb_cases_per_year['new_sp_coh']
})

# For global risk map
tb_burden_by_country = df.groupby(['country', 'iso3'])['new_sp_coh'].sum().reset_index()
tb_burden_by_country.columns = ['country', 'iso3', 'total_new_sp_coh']

# For research insights
mdr_tb_by_country = df.groupby('country')['mdr_coh'].sum().sort_values(ascending=False)
mortality_rate_by_country = df.groupby('country')['Mortality Rate (new_sp)'].mean().sort_values(ascending=False)

st.set_page_config(layout="wide")

st.title('Tuberculosis Surveillance Dashboard')

# Sidebar for navigation
st.sidebar.title('Navigation')
page_selection = st.sidebar.radio(
    'Go to',
    [
        'Global Dashboard',
        'Country Analysis',
        'Early Warning System',
        'Forecasting',
        'Global Risk Map',
        'Research Insights',
        'Data Management',
        'Automated Report Generation' # Added new option
    ]
)

# Main content area based on selection
if page_selection == 'Global Dashboard':
    st.header('Global Dashboard - Aggregated Trends')
    st.write('### Total New TB Cases Per Year')
    st.line_chart(total_tb_cases_per_year.set_index('year'))

    st.write('### Average Treatment Success Rate Per Year')
    st.line_chart(average_treatment_success_rate_per_year.set_index('year'))

elif page_selection == 'Country Analysis':
    st.header('Country-Level Detailed Analysis')

    unique_countries = df['country'].unique()
    # Ensure selected_country is defined for the initial load if it was from previous cell execution
    # If not, set a default or use the first unique country.
    if 'selected_country' not in locals() and 'selected_country' not in globals():
        selected_country = unique_countries[0]

    selected_country_analysis = st.selectbox(
        'Select a Country for Detailed Analysis:',
        unique_countries,
        index=list(unique_countries).index(selected_country) if selected_country in unique_countries else 0
    )

    country_analysis_df = df[df['country'] == selected_country_analysis].copy()

    if not country_analysis_df.empty:
        st.write(f'### TB Trends and Epidemiological Indicators in {selected_country_analysis}')
        fig_country, ax_country = plt.subplots(figsize=(12, 6))
        ax_country.plot(country_analysis_df['year'], country_analysis_df['new_sp_coh'], label='New TB Cases (Cohort)', marker='o')
        ax_country.plot(country_analysis_df['year'], country_analysis_df['Treatment Success Rate (new_sp)'], label='Treatment Success Rate (%)', marker='x')
        ax_country.plot(country_analysis_df['year'], country_analysis_df['Mortality Rate (new_sp)'], label='Mortality Rate (%)', marker='s')
        ax_country.plot(country_analysis_df['year'], country_analysis_df['Treatment Failure Rate (new_sp)'], label='Treatment Failure Rate (%)', marker='d')

        ax_country.set_title(f'TB Trends and Epidemiological Indicators in {selected_country_analysis} Over Time')
        ax_country.set_xlabel('Year')
        ax_country.set_ylabel('Value')
        ax_country.legend()
        ax_country.grid(True)
        plt.xticks(rotation=45)
        st.pyplot(fig_country)
    else:
        st.write('No data available for the selected country.')

elif page_selection == 'Early Warning System':
    st.header('Early Warning System')
    st.write('### Countries with Unusual Increase in TB Cases')
    if not alert_countries.empty: # Check if alert_countries is not empty
        st.dataframe(alert_countries)
    else:
        st.write('No unusual increases detected based on current criteria.')

elif page_selection == 'Forecasting':
    st.header('Forecasting Module')
    st.write('### Predicted Future TB Trends')

    m = Prophet(yearly_seasonality=True)
    m.fit(prophet_df)
    future = m.make_future_dataframe(periods=3, freq='YE')
    forecast_prophet = m.predict(future)

    fig1_st = m.plot(forecast_prophet)
    plt.title('TB Cases Forecast')
    plt.xlabel('Year')
    plt.ylabel('New TB Cases Cohort')
    st.pyplot(fig1_st)

    fig2_st = m.plot_components(forecast_prophet)
    st.pyplot(fig2_st)

elif page_selection == 'Global Risk Map':
    st.header('Global Risk Map')
    st.write('### Global TB Burden by Country')
    fig_map = px.choropleth(
        tb_burden_by_country,
        locations='iso3',
        color='total_new_sp_coh',
        hover_name='country',
        color_continuous_scale=px.colors.sequential.Plasma,
        title='Global TB Burden by Country'
    )
    st.plotly_chart(fig_map)

elif page_selection == 'Research Insights':
    st.header('Automated Research Insights')
    st.write('### Top 5 Countries with Highest Total MDR-TB Cohort')
    st.dataframe(mdr_tb_by_country.head())

    st.write('### Top 5 Countries with Highest Average Mortality Rate (new_sp)')
    st.dataframe(mortality_rate_by_country.head())

elif page_selection == 'Data Management': # New section for data management
    st.header('Data Management')
    st.write('Upload a new dataset or reset to the default dataset.')

    uploaded_file = st.file_uploader("Upload your TB surveillance CSV file", type=["csv"])

    if uploaded_file is not None:
        try:
            st.session_state['df'] = load_and_preprocess_data(uploaded_file=uploaded_file)
            st.success("File successfully uploaded and processed!")
            st.experimental_rerun() # Rerun to update dashboard with new data
        except Exception as e:
            st.error(f"Error processing uploaded file: {e}")

    if st.button('Reset to Default Data'):
        st.session_state['df'] = load_and_preprocess_data(file_path='/content/TB_outcomes_2025-12-13.csv')
        st.success("Dataset reset to default!")
        st.experimental_rerun() # Rerun to update dashboard with default data

elif page_selection == 'Automated Report Generation': # New section for report generation
    st.header('Automated Report Generation')
    st.write('Generate a downloadable report for a selected country.')

    unique_countries_report = df['country'].unique()
    selected_country_report = st.selectbox(
        'Select a Country for Report Generation:',
        unique_countries_report
    )

    report_df = df[df['country'] == selected_country_report].copy()

    if not report_df.empty:
        # Select relevant columns for the report
        report_data = report_df[['year', 'new_sp_coh', 'Treatment Success Rate (new_sp)',
                                   'Mortality Rate (new_sp)', 'Treatment Failure Rate (new_sp)']]
        st.write(f'### Epidemiological Indicators for {selected_country_report}')
        st.dataframe(report_data)

        # Provide download button for CSV
        csv_data = report_data.to_csv(index=False).encode('utf-8')
        st.download_button(
            label="Download Report (CSV)",
            data=csv_data,
            file_name=f"{selected_country_report}_TB_Report.csv",
            mime="text/csv",
        )
    else:
        st.write('No data available for the selected country to generate a report.')
```

```python
import streamlit as st
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
from prophet import Prophet # Import Prophet for forecasting

def load_and_preprocess_data(file_path=None, uploaded_file=None):
    """Loads, cleans, and preprocesses the TB outcomes data."""
    if uploaded_file:
        df_new = pd.read_csv(uploaded_file)
    elif file_path:
        df_new = pd.read_csv(file_path)
    else:
        raise ValueError("Either file_path or uploaded_file must be provided.")

    # Handle missing iso2 values
    df_new['iso2'] = df_new['iso2'].fillna('Unknown')

    # Fill NaN values in numerical columns with 0
    numerical_cols_with_nan = df_new.select_dtypes(include=['float64']).columns
    for col in numerical_cols_with_nan:
        df_new[col] = df_new[col].fillna(0)

    # Create 'date' column
    df_new['date'] = pd.to_datetime(df_new['year'], format='%Y')

    # Calculate epidemiological indicators
    df_new['Treatment Success Rate (new_sp)'] = ((df_new['new_sp_cur'] + df_new['new_sp_cmplt']) / df_new['new_sp_coh'] * 100).fillna(0)
    df_new.loc[df_new['new_sp_coh'] == 0, 'Treatment Success Rate (new_sp)'] = 0

    df_new['Mortality Rate (new_sp)'] = (df_new['new_sp_died'] / df_new['new_sp_coh'] * 100).fillna(0)
    df_new.loc[df_new['new_sp_coh'] == 0, 'Mortality Rate (new_sp)'] = 0

    df_new['Treatment Failure Rate (new_sp)'] = (df_new['new_sp_fail'] / df_new['new_sp_coh'] * 100).fillna(0)
    df_new.loc[df_new['new_sp_coh'] == 0, 'Treatment Failure Rate (new_sp)'] = 0

    # For early warning system
    df_new['rolling_mean_3yr'] = df_new.groupby('country')['new_sp_coh'].rolling(window=3, min_periods=1).mean().reset_index(level=0, drop=True)
    df_new['rolling_std_3yr'] = df_new.groupby('country')['new_sp_coh'].rolling(window=3, min_periods=1).std().reset_index(level=0, drop=True)
    df_new['alert'] = (df_new['new_sp_coh'] > (df_new['rolling_mean_3yr'] + 2 * df_new['rolling_std_3yr']))

    return df_new

# Initialize session state for the DataFrame if not already present
if 'df' not in st.session_state:
    st.session_state['df'] = load_and_preprocess_data(file_path='/content/TB_outcomes_2025-12-13.csv')
    print("Default data loaded and preprocessed into session_state['df']")

df = st.session_state['df']

# Re-calculate derived dataframes using the current df in session state
total_tb_cases_per_year = df.groupby('year')['new_sp_coh'].sum().reset_index()
total_tb_cases_per_year.columns = ['year', 'new_sp_coh']

average_treatment_success_rate_per_year = df.groupby('year')['Treatment Success Rate (new_sp)'].mean().reset_index()
average_treatment_success_rate_per_year.columns = ['year', 'Average Treatment Success Rate (new_sp)']

alert_countries = df[df['alert']][['country', 'year', 'new_sp_coh', 'rolling_mean_3yr', 'rolling_std_3yr', 'alert']]

# For forecasting
prophet_df = pd.DataFrame({
    'ds': pd.to_datetime(total_tb_cases_per_year['year'], format='%Y'),
    'y': total_tb_cases_per_year['new_sp_coh']
})

# For global risk map
tb_burden_by_country = df.groupby(['country', 'iso3'])['new_sp_coh'].sum().reset_index()
tb_burden_by_country.columns = ['country', 'iso3', 'total_new_sp_coh']

# For research insights
mdr_tb_by_country = df.groupby('country')['mdr_coh'].sum().sort_values(ascending=False)
mortality_rate_by_country = df.groupby('country')['Mortality Rate (new_sp)'].mean().sort_values(ascending=False)

st.set_page_config(layout="wide")

st.title('Tuberculosis Surveillance Dashboard')

# Sidebar for navigation
st.sidebar.title('Navigation')
page_selection = st.sidebar.radio(
    'Go to',
    [
        'Global Dashboard',
        'Country Analysis',
        'Early Warning System',
        'Forecasting',
        'Global Risk Map',
        'Research Insights',
        'Data Management',
        'Automated Report Generation' # Added new option
    ]
)

# Main content area based on selection
if page_selection == 'Global Dashboard':
    st.header('Global Dashboard - Aggregated Trends')
    st.write('### Total New TB Cases Per Year')
    st.line_chart(total_tb_cases_per_year.set_index('year'))

    st.write('### Average Treatment Success Rate Per Year')
    st.line_chart(average_treatment_success_rate_per_year.set_index('year'))

elif page_selection == 'Country Analysis':
    st.header('Country-Level Detailed Analysis')

    unique_countries = df['country'].unique()
    # Ensure selected_country is defined for the initial load if it was from previous cell execution
    # If not, set a default or use the first unique country.
    if 'selected_country' not in locals() and 'selected_country' not in globals():
        selected_country = unique_countries[0]

    selected_country_analysis = st.selectbox(
        'Select a Country for Detailed Analysis:',
        unique_countries,
        index=list(unique_countries).index(selected_country) if selected_country in unique_countries else 0
    )

    country_analysis_df = df[df['country'] == selected_country_analysis].copy()

    if not country_analysis_df.empty:
        st.write(f'### TB Trends and Epidemiological Indicators in {selected_country_analysis}')
        fig_country, ax_country = plt.subplots(figsize=(12, 6))
        ax_country.plot(country_analysis_df['year'], country_analysis_df['new_sp_coh'], label='New TB Cases (Cohort)', marker='o')
        ax_country.plot(country_analysis_df['year'], country_analysis_df['Treatment Success Rate (new_sp)'], label='Treatment Success Rate (%)', marker='x')
        ax_country.plot(country_analysis_df['year'], country_analysis_df['Mortality Rate (new_sp)'], label='Mortality Rate (%)', marker='s')
        ax_country.plot(country_analysis_df['year'], country_analysis_df['Treatment Failure Rate (new_sp)'], label='Treatment Failure Rate (%)', marker='d')

        ax_country.set_title(f'TB Trends and Epidemiological Indicators in {selected_country_analysis} Over Time')
        ax_country.set_xlabel('Year')
        ax_country.set_ylabel('Value')
        ax_country.legend()
        ax_country.grid(True)
        plt.xticks(rotation=45)
        st.pyplot(fig_country)
    else:
        st.write('No data available for the selected country.')

elif page_selection == 'Early Warning System':
    st.header('Early Warning System')
    st.write('### Countries with Unusual Increase in TB Cases')
    if not alert_countries.empty: # Check if alert_countries is not empty
        st.dataframe(alert_countries)
    else:
        st.write('No unusual increases detected based on current criteria.')

elif page_selection == 'Forecasting':
    st.header('Forecasting Module')
    st.write('### Predicted Future TB Trends')

    m = Prophet(yearly_seasonality=True)
    m.fit(prophet_df)
    future = m.make_future_dataframe(periods=3, freq='YE')
    forecast_prophet = m.predict(future)

    fig1_st = m.plot(forecast_prophet)
    plt.title('TB Cases Forecast')
    plt.xlabel('Year')
    plt.ylabel('New TB Cases Cohort')
    st.pyplot(fig1_st)

    fig2_st = m.plot_components(forecast_prophet)
    st.pyplot(fig2_st)

elif page_selection == 'Global Risk Map':
    st.header('Global Risk Map')
    st.write('### Global TB Burden by Country')
    fig_map = px.choropleth(
        tb_burden_by_country,
        locations='iso3',
        color='total_new_sp_coh',
        hover_name='country',
        color_continuous_scale=px.colors.sequential.Plasma,
        title='Global TB Burden by Country'
    )
    st.plotly_chart(fig_map)

elif page_selection == 'Research Insights':
    st.header('Automated Research Insights')
    st.write('### Top 5 Countries with Highest Total MDR-TB Cohort')
    st.dataframe(mdr_tb_by_country.head())

    st.write('### Top 5 Countries with Highest Average Mortality Rate (new_sp)')
    st.dataframe(mortality_rate_by_country.head())

elif page_selection == 'Data Management': # New section for data management
    st.header('Data Management')
    st.write('Upload a new dataset or reset to the default dataset.')

    uploaded_file = st.file_uploader("Upload your TB surveillance CSV file", type=["csv"])

    if uploaded_file is not None:
        try:
            st.session_state['df'] = load_and_preprocess_data(uploaded_file=uploaded_file)
            st.success("File successfully uploaded and processed!")
            st.experimental_rerun() # Rerun to update dashboard with new data
        except Exception as e:
            st.error(f"Error processing uploaded file: {e}")

    if st.button('Reset to Default Data'):
        st.session_state['df'] = load_and_preprocess_data(file_path='/content/TB_outcomes_2025-12-13.csv')
        st.success("Dataset reset to default!")
        st.experimental_rerun() # Rerun to update dashboard with default data

elif page_selection == 'Automated Report Generation': # New section for report generation
    st.header('Automated Report Generation')
    st.write('Generate a downloadable report for a selected country.')

    unique_countries_report = df['country'].unique()
    selected_country_report = st.selectbox(
        'Select a Country for Report Generation:',
        unique_countries_report
    )

    report_df = df[df['country'] == selected_country_report].copy()

    if not report_df.empty:
        # Select relevant columns for the report
        report_data = report_df[['year', 'new_sp_coh', 'Treatment Success Rate (new_sp)',
                                   'Mortality Rate (new_sp)', 'Treatment Failure Rate (new_sp)']]
        st.write(f'### Epidemiological Indicators for {selected_country_report}')
        st.dataframe(report_data)

        # Provide download button for CSV
        csv_data = report_data.to_csv(index=False).encode('utf-8')
        st.download_button(
            label="Download Report (CSV)",
            data=csv_data,
            file_name=f"{selected_country_report}_TB_Report.csv",
            mime="text/csv",
        )
    else:
        st.write('No data available for the selected country to generate a report.')
```
